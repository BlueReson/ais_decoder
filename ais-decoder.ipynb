```python
files = os.listdir(path)
for f in files:
	print(f)
```

```python
parameters =['mmsi', 'nav_status', 'sog', 'position_accuracy', 'y', 'x','cog', 'true_heading', 'timestamp']
```

```python
import csv

ship_dick = {}
for this_file in files[1:]: # cancel out the first line
    file = open(path+'/'+this_file, encoding='utf-8')
    ship_list = []
    for line in file.readlines():
        try:
            message = line.split('"')[1]
            timestamp = line.split('"')[0]
            if message[0] == '!':
                row_data = []
                ais_coded = message.split(',')[5]
                q = ais.decode(ais_coded, 0)
#                 print(q)
                heading = q['true_heading']
    
                if int(heading) != 511:
                    # Add timestamp to the first collomn of the csv 
                    # The time stamp is lightly in a UNIX time format
                    row_data.append(timestamp)
                    
                    #Get all the perameters from the decoded AIS
                    for parameter in parameters:
                        row_data.append(q[parameter])
                    #Lets save the row_data into a csv
                    with open('headingships.csv', 'a+', newline='')  as file:
                        writer = csv.writer(file)
                        writer.writerow(row_data)

                    print(f"{heading} ------ {q['mmsi']}")
#                     print(q)
                    #Make a list of MMSI with heading
                    ship_list.append(q['mmsi'])
        except Exception as e:
            print(f"Error {e}")
        ship_dick[this_file] = ship_list
#     ais_str = str(line[20:])
#     print(ais_str)
#     try:
#         msg = pynmea2.parse(ais_str)
# #         print(repr(msg))
#         print(msg.speed)
#     except pynmea2.ParseError as e:
#         print('Parse error: {}'.format(e))
#         continue
```
